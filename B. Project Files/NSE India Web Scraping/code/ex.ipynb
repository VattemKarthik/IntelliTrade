```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir="D:\\programs\\ads_flask\\logres_op\\90_day_pred"
hist_dir="D:\\programs\\ads_flask\\logres_op\\data"
```

```python
ls_d=[]
files=[]
for root, dirs, files in os.walk(data_dir, topdown=False):
   for name in files:
      ls_d.append(os.path.join(root, name))
for i in range(len(files)):
   files[i]=files[i].replace(".csv","")
```

```python
ls_d[0]
```

```python
ls_h=[]
files=[]
for root, dirs, files in os.walk(hist_dir, topdown=False):
   for name in files:
      ls_h.append(os.path.join(root, name))
for i in range(len(files)):
   files[i]=files[i].replace(".csv","")
```

```python
# df=pd.DataFrame(columns=["current","estimated","pct_change"])
df=[]
for i in range(len(files)):
    d=pd.read_csv(ls_d[i])
    h=pd.read_csv(ls_h[i])
    df.append([files[i],d["predicted value"][89],h["close "][h["OPEN "].size-1],(d["predicted value"][89]-h["close "][h["OPEN "].size-1])*100/h["close "][h["OPEN "].size-1]])
```

```python
df
```

```python
df=pd.DataFrame(df,columns=["name","20/7/2023 price","t+90 Day price","pct change"])
```
