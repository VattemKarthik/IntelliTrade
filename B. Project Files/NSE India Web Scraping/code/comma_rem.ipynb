```python
import pandas as pd
import numpy as np
import os
```

```python
# removes the annoying comma preent in stock data
```

```python
read_data_dir="D:\\programs\\Applied data Science\\proj\\output"
data_write_dir="D:\\programs\\Applied data Science\\proj\\cleaned"
symbols_read_path="D:\\programs\\Applied data Science\\MW-NIFTY-50-symbols.csv"
```

```python
stocks=pd.read_csv(symbols_read_path)
stocks
```

```python
ls=[]
files=[]
for root, dirs, files in os.walk(read_data_dir, topdown=False):
   for name in files:
      ls.append(os.path.join(root, name))
```

```python
j=0
for i in ls:
    df=pd.read_csv(i)  
    df=df.replace({',':''},regex=True)
    # print(df)
    df=df.drop(["Unnamed: 0"],axis=1)
    df.to_csv(path_or_buf=data_write_dir+"\\"+files[j])
    j+=1
```
